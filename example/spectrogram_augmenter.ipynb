# Visual Original Spectrogram

```python
from nlpaug.util.file.load import LoadUtil
from nlpaug.util.visual.spectrogram import VisualSpectrogram

path = 'Yamaha-V50-Rock-Beat-120bpm.wav'

mel_spectrogram = LoadUtil.load_mel_spectrogram(path, n_mels=128)
VisualSpectrogram.visual('Original', mel_spectrogram)
```

```python
import nlpaug.augmenter.spectrogram as nas
import nlpaug.flow as naf
```

# Frequency Masking

```python
aug = nas.FrequencyMaskingAug(mask_factor=80)

augmented_mel_spectrogram = aug.substitute(mel_spectrogram)
VisualSpectrogram.visual('Augmented', augmented_mel_spectrogram)
```

# Time Masking

```python
aug = nas.TimeMaskingAug(mask_factor=80)

augmented_mel_spectrogram = aug.substitute(mel_spectrogram)
VisualSpectrogram.visual('Augmented', augmented_mel_spectrogram)
```

# Combine Frequency Masking and Time Masking

```python
import nlpaug.flow as naf

flow = naf.Sequential([
    nas.FrequencyMaskingAug(mask_factor=50), 
    nas.TimeMaskingAug(mask_factor=20), 
    nas.TimeMaskingAug(mask_factor=30)])
augmented_mel_spectrogram = flow.augment(mel_spectrogram)
VisualSpectrogram.visual('Augmented', augmented_mel_spectrogram)
```
